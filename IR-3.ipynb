```python
import numpy as np
import pandas as pd
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
```

```python
df=pd.read_csv('processed_cleveland.csv')
```

```python
df=df.replace('?',np.nan)
```

```python
df.head()
```

```python
model=BayesianNetwork([('age','trestbps'),('age','fbs'),('sex','trestbps'),('exang','trestbps'),('trestbps','num'),
                     ('fbs','num'),('num','restecg'),('num','thalach'),('num','chol')])
```

```python
print('\nLearning CPD using Bayesian Network:')
model.fit(df,estimator=MaximumLikelihoodEstimator)

print("Inferencing with Bayesian Network:")
df_infer=VariableElimination(model)
print("1.Probability with age 38")
q=df_infer.query(variables=['num'],evidence={'age':38})
print(q)
print("1.Probability with cholestrol 230")
p=df_infer.query(variables=['num'],evidence={'chol':230})
print(p)
```
